## Functions

```python
def add_pos_tags(text:str) -> str:
    """Add POS-tags to text (format: "my_DET example_NOUN").

    Args:
        text (str): text without POS-tags

    Returns:
        str: text with POS-tags
    """
    tokens_with_pos = []
    for token in nlp(text):
        tokens_with_pos.append(f"{token}_{token.pos_}")
    return " ".join(tokens_with_pos)

def add_pos_with_context(series:pd.Series, pre_context:str, post_context:str="") -> pd.Series:
    """Add POS tags to text in series. The tags are assigned based on the word's role in the provided context.

    Args:
        series (pd.Series): the series of text without POS tags
        pre_context (str): the sentence before the word from the series, e.g. "He gave" if the context is: "He gave {word} to her".
        post_context (str, optional): the sentence after the word, e.g. "to her" if the context is: "He gave {word} to her". Defaults to "".

    Returns:
        pd.Series: the series of text with POS tags
    """
    res = [] # store results
    text_idx = len(pre_context.split()) # text is located after pre_context
    for text in series:
        text = text.lower()
        text_with_tags = add_pos_tags(pre_context + " " + text + " " + post_context).split()
        res.append(" ".join( text_with_tags[ text_idx : text_idx+len(text.split()) ] ))
    return pd.Series(res)

def remove_pos_tags(text:str) -> str:
    """Remove POS-tags from text (e.g.: "example_NOUN" -> "example").

    Args:
        text (str): text with POS-tags, e.g. "my_DET example_NOUN"

    Returns:
        str: text without POS-tags, e.g. "my example"
    """
    pattern = re.compile(r"_([A-Z]+)\b")
    return pattern.sub("", text)

def remove_punctuation(string:str) -> str:
    """Remove punctuation from a string.
    
    Args: 
        string (str): string with punctuation
    
    Returns:
        str: string without punctuation
    """
    return string.translate(str.maketrans("", "", punctuation))

def pick_substitution(orig_word:str, gender_triplet:Dict[str, str], gender_2_gender:Dict[str, str]) -> str:
    """If the word is in the gender dictionary, find another word to substitute it with (one of the two other gender categories).

    Args:
        orig_word (str): the original word.
        gender_dict (Dict[str, str]): dictionary of the gender triplets, i.e. {female: female_word, male: male_word, queer: queer_word}.
        gender_2_gender (Dict[str, str]): if the gender associated with a word is that of the key, it will be substituted with the corresponding value.

    Returns:
        str/None: the new word or None if the word is not gendered.
    """
    for gender, gender_word in gender_triplet.items():
        if gender_word == orig_word:
            new_gender = gender_2_gender[gender] # extract new gender category
            new_word = gender_triplet[new_gender] # extract new word in that gender category            
            return new_word
    return None # if there are no matches
```

## Perform CDS
Takes approx. 10 minutes to run.

```python
# example of within-document name anonymization 
# Larsen should be given the same tag within documents, but different tags across documents (same for Kim)

ex = pd.DataFrame(["En sætning om Kim Larsen. Det er sandt, at Kim elsker kaffe.", "Nu handler det om Kim Petras", "Hr. Larsen giver fru Larsen en buket blomster én gang om året.", "!!!"], columns=["text"])

names_counter = 0

for text in ex["text"]:
    
    # anonymize names and add POS tags
    name_anonymous_dict = {}
    text = remove_punctuation(text.lower())
    text_pos = "" #text with POS-tags
    if len(text) > 0: 
        print(text)
        doc = nlp(text)
        for token in doc:
            token_str = str(token)
            if token.ent_type_ == "PER":
                
                if token_str not in name_anonymous_dict:
                    names_counter += 1
                    name_anonymous_dict[token_str] = "E"+str(names_counter)
                token_str = name_anonymous_dict[token_str] 
                print(f"   {str(token).ljust(7)} →  {token_str}")
                # print(token, token.pos_, token_str)
                # text_pos += " "+token_str+"_PROPN" # manually assign POS-tag as it's a string, not token object
            text_pos += f" {token_str}_{token.pos_}" 
    print()
```

```python
# initialize variables
new_train_X = [] # for storing new training data
names_counter = 0 # count unique names
names = [] # store all names
new_nouns = [] # store new nouns
old_nouns = [] # store old nouns
skip_counter = 0 # count number of skipped documents

for text in tqdm(train_orig["tweet"]):

    action = choices(["skip", "swap"], weights=[1/3, 2/3], k=1)[0] # pick whether to skip or apply substitutions to this document
    if action == "swap": 
        gender_2_gender = { # pick the gender category that original gender should be flipped to
            "female":choice(["male", "queer"]),
            "male":choice(["female", "queer"]),
            "queer":choice(["female", "male"])
            }
    elif action == "skip":
        skip_counter += 1

    name_anonymous_dict = dict() # store anonymization tags

    # 1) remove punctuation and lowercase
    text = remove_punctuation(text.lower())
    text_w_pos = "" # text with pos-tags
    
    if len(text) > 0: # if string not empty (else, an empty string is added to text_w_pos at the end)
        doc = nlp(text)
        for token in doc:
            token_str = str(token) # string version of token
            
            # 2) anonymize the names in the text and add POS tags 
            if token.ent_type_ == "PER":
                if token_str not in name_anonymous_dict: # if not already in dict, add it
                    names_counter += 1
                    name_anonymous_dict[token_str] = "E"+str(names_counter)
                token_str = name_anonymous_dict[token_str] # make substitution
                names.append(token_str)
            token_w_pos = f"{token_str}_{token.pos_}"
            text_w_pos += " "+token_w_pos

            if action == "swap":
                # 3) substitute the gendered words with other gendered words
                for idx in gender_dict_with_POS:
                    new_word = pick_substitution(token_w_pos, gender_dict_with_POS[idx], gender_2_gender)
                    if new_word is not None: # i.e. if substitution should be performed
                        text_w_pos = text_w_pos.replace(token_w_pos, new_word)
                        old_nouns.append(token_w_pos)
                        new_nouns.append(new_word)

            # 4) remove pos tag
            text_wo_pos = remove_pos_tags(text_w_pos)

    new_train_X.append(text_wo_pos)

print(f"Skipped {skip_counter} documents ({(skip_counter/len(new_train_X))*100:.2f}%)")
```

## Frequencies before CDS

Approach:
* on training data:
    * remove punctuation from lowercased text
    * add pos tags
    * make simple frequency dictionary
    * for each gender category (female, male, queer/other):
        * for each word in that category:
            * add relative frequency to new frequency dict if it's above 0
    * plot relative frequencies per gender category (only those that appear at least once)

```python
# preprocess and add pos-tags
train_orig_pos = []

for text in tqdm(train_orig["tweet"]): 
    text = remove_punctuation(text.lower())
    text_w_pos = ""
    if len(text) > 0:
        doc = nlp(text)
        for token in doc:
            token_w_pos = f"{token}_{token.pos_}"
            text_w_pos += " "+token_w_pos
        train_orig_pos.append(text_w_pos)

X_train_orig_pos = pd.Series(train_orig_pos)
```

```python
X_train_orig_pos
```

```python
orig_freqs = {}
for text in X_train_orig_pos:
    for word in text.split():
        orig_freqs[word] = orig_freqs.get(word, 0)+1

orig_n_words = sum(orig_freqs.values())
print(f"{orig_n_words:,} words in total")
print(f"{len(orig_freqs.keys()):,} unique words")
```

```python
gender_df_with_POS.head()
```

```python
female_freqs = {}
female_no_occ = []
for female_word in gender_df_with_POS["female"]:
    freq = orig_freqs.get(female_word, 0)
    if freq > 0:
        female_freqs[female_word] = freq / orig_n_words
    else:
        female_no_occ.append(female_word)
sorted_female_freqs = dict(sorted(female_freqs.items(), key=lambda x:x[1], reverse=True))

male_freqs = {}
male_no_occ = []
for male_word in gender_df_with_POS["male"]:
    freq = orig_freqs.get(male_word, 0)
    if freq > 0:
        male_freqs[male_word] = freq / orig_n_words
    else:
        male_no_occ.append(male_word)
sorted_male_freqs = dict(sorted(male_freqs.items(), key=lambda x:x[1], reverse=True))

queer_freqs = {}
queer_no_occ = []
for queer_word in gender_df_with_POS["queer"]:
    freq = orig_freqs.get(queer_word, 0)
    if freq > 0:
        queer_freqs[queer_word] = freq / orig_n_words
    else:
        queer_no_occ.append(queer_word)
sorted_queer_freqs = dict(sorted(queer_freqs.items(), key=lambda x:x[1], reverse=True))
```

```python
print("Number of terms that don't appear whatsoever:")
print(f"Female: {len(female_no_occ)}  ({(len(female_no_occ)/197)*100:.2f}%)")
print(f"Male  : {len(male_no_occ)}  ({(len(male_no_occ)/197)*100:.2f}%)")
print(f"Queer : {len(queer_no_occ)} ({(len(queer_no_occ)/197)*100:.2f}%)")
```

```python
# plot results
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot the first subplot
axs[0].scatter(x=list(sorted_female_freqs.keys()), y=list(sorted_female_freqs.values()), zorder=2, color="b")
axs[0].set_title('Female')
axs[0].tick_params(axis='x', rotation=90)
axs[0].set_ylim([0, 0.005])
axs[0].grid(zorder=1)

# Plot the second subplot
axs[1].scatter(x=list(sorted_male_freqs.keys()), y=list(sorted_male_freqs.values()), zorder=3, color="g")
axs[1].set_title('Male')
axs[1].tick_params(axis='x', rotation=90)
axs[1].set_ylim([0, 0.005])
axs[1].grid(zorder=1)

# Plot the third subplot
axs[2].scatter(x=list(sorted_queer_freqs.keys()), y=list(sorted_queer_freqs.values()), zorder=4, color="m")
axs[2].set_title('Other')
axs[2].tick_params(axis='x', rotation=90)
axs[2].set_ylim([0, 0.005])
axs[2].grid(zorder=1)

# Adjust layout to prevent overlapping
fig.suptitle("Relative frequency of gendered words in the original training set")
plt.tight_layout()

# Show the plots
plt.savefig(os.getcwd()+"/plots/freq_b4_CDS.png")
plt.show()
```

## Frequencies after CDS

```python
# preprocess and add pos-tags
train_CDS_pos = []
no_occurrences = []
for text in tqdm(train_new["text"]): 
    text = remove_punctuation(text.lower())
    text_w_pos = ""
    if len(text) > 0:
        doc = nlp(text)
        for token in doc:
            if str(token) != " ":
                token_w_pos = f"{token}_{token.pos_}"
                text_w_pos += " "+token_w_pos
        train_CDS_pos.append(text_w_pos)
    else:
        no_occurrences += text
        
X_train_CDS_pos = pd.Series(train_CDS_pos)
```

```python
X_train_CDS_pos
```

```python
CDS_freqs = {}
for text in X_train_CDS_pos:
    for word in text.split():
        CDS_freqs[word] = CDS_freqs.get(word, 0)+1

CDS_n_words = sum(CDS_freqs.values())
print(f"{CDS_n_words:,} words in total")
```

```python
female_freqs = {}
female_no_occ = []
for female_word in gender_df_with_POS["female"]:
    freq = CDS_freqs.get(female_word, 0)
    if freq > 0:
        female_freqs[female_word] = freq / CDS_n_words
    else:
        female_no_occ.append(female_word)
sorted_female_freqs = dict(sorted(female_freqs.items(), key=lambda x:x[1], reverse=True))

male_freqs = {}
male_no_occ = []
for male_word in gender_df_with_POS["male"]:
    freq = CDS_freqs.get(male_word, 0)
    if freq > 0:
        male_freqs[male_word] = freq / CDS_n_words
    else:
        male_no_occ.append(male_word)
sorted_male_freqs = dict(sorted(male_freqs.items(), key=lambda x:x[1], reverse=True))

queer_freqs = {}
queer_no_occ = []
for queer_word in gender_df_with_POS["queer"]:
    freq = CDS_freqs.get(queer_word, 0)
    if freq > 0:
        queer_freqs[queer_word] = freq / CDS_n_words
    else:
        queer_no_occ.append(queer_word)
sorted_queer_freqs = dict(sorted(queer_freqs.items(), key=lambda x:x[1], reverse=True))
```
